###  1. agregar_producto(nombre, precio, cantidad) : Agrega un producto al inventario o
actualiza su cantidad si ya existe. Debe recibir el nombre, precio y cantidad del producto como
parámetros.

```python
def agregar_producto(nombre, precio, cantidad):
    for producto in inventario_zapateria:
      if producto["nombre"] == nombre:
         producto["cantidad"] += cantidad
         producto["precio"] = precio  # opcional: actualizar precio si cambia
         print("El producto está en el inventario, y hemos actualizado la cantidad")
         return
    inventario_zapateria.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    print("El producto es nuevo, lo hemos añadido")    
```

# 2. ver_inventario(): Muestra el inventario de productos con sus detalles.

## Utiliza un bucle for para recorrer el inventario. Compara los nombres de los productos con el nombre proporcionado. Si se encuentra el producto, imprime sus detalles.

```python
def ver_inventario(nombre_buscar):
    for producto in inventario_zapateria:
        if producto['nombre'] == nombre_buscar:
            if producto['cantidad'] != 0:
                print(f"producto: {producto['nombre']}, precio : {producto['precio']}, cantidad: {producto['cantidad']}")
            else:
                print(f"producto {producto['nombre']} agotado")   
            return 
    print(f"no se encuentra el producto {nombre_buscar} en el inventario")
              
    
```

##  4. actualizar_stock(nombre, cantidad): Actualiza el stock de un producto en el inventario.
Debe recibir el nombre del producto y la cantidad a agregar o quitar como parámetros.

```python
def actualizar_stock(nombre, cantidad):
    for producto in inventario_zapateria:
        if producto.get("nombre") == nombre:
            producto["cantidad"] = producto.get("cantidad", 0) + cantidad
            print("El producto está en el inventario, y hemos actualizado la cantidad")
            return producto
    print("El producto no está en el inventario")
    return
```

```python
actualizar_stock("chancla_playa", 5)
```

```python
actualizar_stock("bota_nieve", 10)
```

##  5. eliminar_producto(nombre): Elimina un producto del inventario por nombre. Debe recibir el nombre del producto como parámetro.

###  Utiliza un bucle for para recorrer el inventario. Busca el producto por nombre. Elimina el producto del inventario si existe. Si el producto no esta en el inventario muestra un mensaje indicándolo.

```python
def eliminar_producto(nombre):
    for producto in (inventario_zapateria):
        if producto.get("nombre") == nombre:
            inventario_zapateria.remove(producto)
            print("El producto ha sido eliminado del inventario")
            return
    print("El producto no está en el inventario")
```

```python
eliminar_producto("zapato_tacon")
```

```python
eliminar_producto("bota_sandalia")
```

## 6. calcular_valor_inventario(): Calcula y muestra el valor total del inventario.

###  Utiliza un bucle for para calcular el valor total del inventario. Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir, calcula el valor total del inventario

```python
def calcular_valor_inventario():
    valor_total = 0
    for producto in inventario_zapateria:
        valor_total += producto["precio"] * producto["cantidad"]
    return valor_total
```

```python
calcular_valor_inventario()
```

##  7. realizar_compra(): Permite a un cliente realizar una compra seleccionando productos del inventario. Debe interactuar con el cliente para seleccionar productos y calcular el costo total de la compra.

```python
def realizar_compra():
    while True:
        global ventas_totales
  
        nombre_producto = input("Ingrese el nombre del producto a comprar: ")
        try:
            cantidad = int(input("Ingrese la cantidad a comprar: "))
        except ValueError:
            print("Cantidad inválida")
            return
        for producto in inventario_zapateria:
            if producto.get("nombre") == nombre_producto:
                if producto.get("cantidad", 0) >= cantidad:
                    producto["cantidad"] -= cantidad
                    total_compra = producto["precio"] * cantidad
                    ventas_totales += total_compra
                    print(f"Compra realizada: {cantidad} x {nombre_producto} = Total: {total_compra}")
                    return
                else:
                    print("No hay suficiente stock para completar la compra")
                    return
        print("El producto no está en el inventario")
        break   
```

```python
print(inventario_zapateria)
```

```python
realizar_compra()
```
